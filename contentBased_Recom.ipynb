```python
# Create a vector list for whole user with to movie vector

vector=[]
for user in range(1,944):
    u1=ratings[ratings['user_id']==user][['movie_id','rating']]
    u1_list=[]
    for i in u1['movie_id']:
        f=movie_vector[i].values
        u1_list.append(f)
    uu1=pd.DataFrame(u1_list,columns=range(1,19))
    
    vec=[]
    for j in range(1,19):
        g=sum(uu1[j])
        vec.append(g/272)
    vector.append(vec)
```

```python
# function

def user_vector(ratings,movie_vector):
    vector=[]
    for user in range(1,944):
        u1=ratings[ratings['user_id']==user][['movie_id','rating']]
        u1_list=[]
        for i in u1['movie_id']:
            f=movie_vector[i].values
            u1_list.append(f)
        uu1=pd.DataFrame(u1_list,columns=range(1,19))
        
        vec=[]
        for j in range(1,19):
            g=sum(uu1[j])
            vec.append(g/272)
        vector.append(vec)
    return vector
```

```python
vector=user_vector(ratings,movie_vector)
```

```python
users_vector=pd.DataFrame(vector)
```

```python
users_vector
```

```python
users_vector.shape
```

```python
movie_vector.shape
```

```python
# Dot matrix for  user_vector and movie_vector
content_vector=np.dot(users_vector,movie_vector)
```

```python
# Data Frame  for content vector
content_vector=pd.DataFrame(content_vector)
```

```python
# Transpose for content vector
content_vectors=content_vector.T
```

```python
def content_based(ratings,content_vectors,user,thresh):
    u2=ratings[ratings['user_id']==user]['movie_id'].values
    val=content_vectors[content_vectors[user]>(content_vectors[user].max()-thresh)][2]
    recom=[]
    for i in val.index:
        if(i in u2):
            pass
        else:
            recom.append(i)

    return recom
```
